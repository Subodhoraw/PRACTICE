```python
5️⃣ Clean phone numbers in JSON: remove spaces, +91, dashes
```

```python
import json 
import re # regular expressions the best toll tp pattern matching
messy_numbers = [
    {"name": "Rahul", "phone": "+91-9876543210"},  # Dash and Code
    {"name": "Sara",  "phone": "987 654 3211"},    # Spaces
    {"name": "John",  "phone": "(+91)9876543212"}, # Parentheses
    {"name": "Amit",  "phone": "987-654-3213"},    # Multiple Dashes
    {"name": "Priya", "phone": "+91 98765 43214"}, # Code + Space
    {"name": "Mike",  "phone": "9876543215"},      # Already Clean
    {"name": "Anuj",  "phone": "UNKNOWN"},         # Garbage data
    {"name": "Lisa",  "phone": "09876543216"}      # Leading Zero
]

print("----1. Messy data created ---")
 #print(json.dumps(messy_numbers,indent =2 )) #indent

def clean_numbers(phone_str):
    if not phone_str or not any(char.isdigit() for char in phone_str):
        return None
    #remove specefic prefixes(like +91 or 0 at start)
    #we remove +91 or 0 that start with
    phone_str = phone_str.replace("+91", "")
    #3. remove all non numeric characters(spaces, dashes, brackets)
    #\D means "anything that is not a digit"
    clean_num = re.sub(r'\D','',phone_str)

    #remove leading zer if present (standardizing to 10 digits)
    if clean_num.startswith('0'):
       clean_num = clean_num[1:]
    return clean_num
cleaned_data = []

for entry in messy_numbers:
    original = entry["phone"]
    clean  = clean_numbers(original)

    # only add successfully processed data
    if clean:
        cleaned_data.append({
            "name ": entry["name"],
            "original_phone": original,
            "clean_phone" : clean
        })

# save to json 

output_file = "cleaned_num.json"
with open(output_file,'w') as f:
    json.dump(cleaned_data,f,indent = 4)

print(f"data cleaned and savded to folder {output_file}")
print(json.dumps(cleaned_data, indent = 4))


```

## duplicate rows in a CSV and print them

(Real-world ETL prblem)

```python
import pandas as pd
data = {
    'CustomerID': [101, 102, 103, 101, 104, 102, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Bob', 'Eve'],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 
              'alice@example.com', 'david@example.com', 'bob@example.com', 'eve@example.com'],
    'SignupDate': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-01', 
                   '2023-01-04', '2023-01-02', '2023-01-05']
}

# create the file 
df_raw = pd.DataFrame(data)
df_raw.to_csv('customer_with_dumps.csv', index = False)
print("created 'customer_with_dups.csv")
# indentify dupplicates 
# keep = 
#================================

#2 ETL solution
#==============================
#1. load data
df = pd.read_csv('customer_with_dumps.csv')

#2. identify  duplicates 
# keep='first'  -> Marks the 2nd/3rd instance as True (standard)
# keep=False    -> Marks ALL instances as True (useful to see the groups)

duplicate_rows = df[df.duplicated(keep = False)]

#3. Print the results 
if not duplicate_rows.empty:
    print(f" found {len(duplicate_rows)}   duplicate rows ")
    print(duplicate_rows)
#optional save in diffrent folder for audit 
    duplicate_rows.to_csv("audit_duplicates.csv", index = False)
    print("\n(duplicates saved to 'audit_duplicates.csv' for review)")
else:
    print('\n no duplicates found in csv file')
```
